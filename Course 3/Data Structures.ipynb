```python
def to_dollars_cents(price):
    ''' 
    Split price into dollars and cents
    '''
    dollars = int (price // 1 ) 
    cents = round ((price % 1 ) * 100)
    return dollars, cents
```

```python
to_dollars_cents( 7.99)
```

```python
dollar , cent = to_dollars_cents( 10.59)
print(dollar)
print(cent)
```

```python
#Unpack the tuple extract name and player position
team = [
    ('Marta', 20, 'center'),
    ('Karthika', 30, 'point guard'),
    ('Sanjan', 7, 'power forward'),
    ('Mithran', 4, 'small forward'),
    ('Murali', 32, 'shooting guard')
]
```

```python
def player_position(players):
    result = []
    for name, age, position in players:
        result.append('Name: {:>20} \nPosition: {:>16}\n'.format(name,position))

    return result
        
```

```python
for player in player_position(team):
    print(player)
```

# Convert tuple into dictionary

```python
team = [
    ('Marta', 20, 'center'),
    ('Karthika', 30, 'point guard'),
    ('Sanjan', 7, 'power forward'),
    ('Mithran', 4, 'small forward'),
    ('Murali', 32, 'shooting guard'),
    ('Lucy', 20, 'center'),
    ('Violet', 30, 'point guard'),
    ('Alex', 7, 'power forward'),
    ('Chase', 4, 'small forward'),
    ('Luca', 32, 'shooting guard')
]

new_team = {}
for name,age,position in team:
    if position in new_team:
        new_team[position].append((name,age))
    else:
        new_team[position] = [(name, age)]
```

```python
# below one throws error because each element of a set must be hashable (≈ immutable), but {‘a’, ‘b’, ‘c’} is a set, which is a mutable 
example_c = [1.5, {'a', 'b', 'c'}, 1.5]
set(example_c)
```

```python
#intersection
set1 = {1, 2, 3, 4, 5, 6}
set2 = {2, 3, 6, 7, 8}
print(set1.intersection(set2))
print(set1 & set2)
```

```python
# union
set_1 = {'a', 'b', 'c'}
set_2 = {'b', 'c', 'd'}
set_1.union(set_2)
set_1 | set_2
```

```python
#difference
print(set_1.difference(set_2))
print(set_1 - set_2)
print(set_1.symmetric_difference(set_2))
print(set_1 ^ set_2)
```

```python
import pandas as pd

df4 = pd.DataFrame({
    'Planet': [
        'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune',
        'Janssen', 'Tadmor'
    ],
    'type': [
        'Terrestrial', 'Terrestrial', 'Gas Giant', 'Gas Giant',
        'Ice Giant', 'Ice Giant',
        'Exoplanet', 'Exoplanet'
    ],
    'rings': [
        False, False, True, True, True, True,
        False, False
    ],
    'mean_temp_c': [
        15, -65, -110, -140, -195, -200,
        1700, -50   # approximate exoplanet values
    ],
    'magnetic_field': [
        True, False, True, True, True, True,
        False, False   # unknown → False by default
    ],
    'life': [
        True, False, False, False, False, False,
        False, False
    ]
})

df4
```

```python
# inner join (planets common in both dataframe)
inner = pd.merge(df3, df4, on = 'Planet', how = 'inner')
inner
```

```python
# outer join (all planets  in both dataframe)
outer = pd.merge(df3, df4, on = 'Planet', how = 'outer')
outer
```

```python
# left join (all planets  in left dataframe along with corresponding column values in right dataframe)
left = pd.merge(df3, df4, on = 'Planet', how = 'left')
left
```

```python
# Right join 
right = pd.merge(df3, df4, on = 'Planet', how = 'right')
right
```
